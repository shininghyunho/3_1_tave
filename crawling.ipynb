```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
def remove_tag(content):
   cleanr =re.compile('<.*?>')
   cleantext = re.sub(cleanr, '', content)
   return cleantext
```

```python
remove_tag('<h3 class="petitionsView_title">세종대왕 탄신일(5월15일)을 국가기념일로 지정해 주셨으면 합니다.</h3>')
```

```python
pe_article_id=[]
pe_start = []
pe_end = []
pe_votes = []
pe_categorys = []
pe_titles = []
pe_contents = []
#pe_article_id.append('article_id')
#pe_start.append('start')
#pe_end.append('end')
#pe_votes.append('vote')
#pe_titles.append('title')
#pe_contents.append('content')
for i in range(580200, 580250): 
    URL = "https://www1.president.go.kr/petitions/"+str(i)
    response = requests.get(URL)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    titles = soup.find_all("h3",{"class" :"petitionsView_title"})
    votes = soup.find_all("span", {"class" : "counter"})
    contents = soup.find_all("div", {"class" : "View_write"})
    list_tag=[]
    for ultag in soup.find_all('ul', {'class': 'petitionsView_info_list'}):
        for litag in ultag.find_all('li'):
            #print(litag.text)
            list_tag.append(litag.text)
            #print("--------")
            
    #print(i)
    #print(list_tag[0][4:])
    #print(list_tag[1][4:])
    #print(list_tag[2][4:])
    #print(type(titles[0]))
    #print(titles[0])
    #titles_str=titles[0].string
    #print(type(titles_str))
    #print(titles_str)
    #print(remove_tag(titles_str))
    #print(remove_tag((str)titles[0]))
    #print(remove_tag((str)contents[0]))

    
    #print(titles[0])
    #print(votes[0].string)
    #print(contents[0])
    #print(contents[0])
    #print(titles[0])
    #print(contents[0].get_text(" ", strip=True)[:-58])
    
    pe_article_id.append(i)
    pe_categorys.append(list_tag[0][4:])
    pe_start.append(list_tag[1][4:])
    pe_end.append(list_tag[2][4:])
    pe_titles.append(titles[0].string)
    pe_contents.append(contents[0].get_text(" ", strip=True)[:-58])
    pe_votes.append(votes[0].string)
    
    
    
  
        
#e[5:8]
#e[12:22]
#e[27:37]

#contents = soup.find_all("h3",{"class" : "petitionsView_title"})
#contents
 
```
