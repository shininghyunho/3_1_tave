```python
text='A범죄자는 제발 사형시켜주세요. 아주 나쁜놈입니다. 게다가 짱깨입니다. 청와대가 직접 수사해 반드시 처벌 바랍니다.'
```

```python
# 1. 중복체크
if len(df.loc[df["content"]==text])>0:
    print("중복된 청원이 존재합니다.")
    print(df.loc[df["content"]==text]["content"])
else:
    print("중복된 청원이 존재하지 않습니다.")
```

```python
malist = twitter.pos(text, norm=True, stem=True)
```

```python
results=[]
for word in malist:
            # 어미/조사/구두점 등은 대상에서 제외 
            if not word[1] in ["Josa", "Eomi", "Punctuation"]:
                results.append(word[0])
```

```python
type(results)
```

```python
results
```

```python
# 2-1. 비속어 먼저 체크
import re
p = re.compile('.*(개새끼|쌍놈|새끼|미친|병신|미치다|문죄인|문재앙|빨갱이|존나|놈|등신|좆|좃).*')
bad=[]
bad.clear()
for mal in results:
    #print(mal)
    for r in p.findall(mal):
        bad.append(r)
if len(bad)>0:
    print("비속어가 검출되었습니다.")
    print("비속어가 맞는지 확인해주세요.")
    print(bad)
else:
    print("비속어가 검출되지 않았습니다.")
```

```python
# 2-2. 폭력적 어휘
import re
p = re.compile('.*(정신병자|빌어먹을|죽여주다|죽이다).*')
bad=[]
bad.clear()
for mal in results:
    #print(mal)
    for r in p.findall(mal):
        #print(r)
        bad.append(r)
if len(bad)>0:
    print("폭력적인 어휘가 검출되었습니다.")
    print("폭력적인 어휘가 맞는지 확인해주세요.")
    print(bad)
else:
    print("폭력적인 어휘가 검출되지 않았습니다.")
```

```python
# 2-2. 집단혐오 어휘
import re
p = re.compile('.*(꼴페미|틀딱|한남|흑형|짱깨|김치녀).*')
bad=[]

for mal in results:
    #print(mal)
    for r in p.findall(mal):
        #print(r)
        bad.append(r)
        
if len(bad)>0:
    print("집단혐오 어휘가 검출되었습니다.")
    print("집단혐오 어휘가 맞는지 확인해주세요.")
    print(bad)
else:
    print("집단혐오 어휘가 검출되지 않았습니다.")
```

```python
# 2-3. 청와대 권한을 넘어서는 어휘
import re
p = re.compile('.*(처벌|사형|파면|반대|억울함|억울|억울하게|폐지|폐간).*')
bad=[]

for mal in results:
    #print(mal)
    for r in p.findall(mal):
        #print(r)
        bad.append(r)
        
if len(bad)>0:
    print("청와대 권한을 넘어서는 어휘가 검출되었습니다.")
    print("재판이 진행 중이거나, 입법부,사법부의 고유 권한과 관련 한 내용으로 삼권분립의 정신을 훼손할 소지가 있는 청원에는 답변이 어려울 수 있습니다..")
    print(bad)
else:
    print("청와대 권한을 넘어서는 어휘가 검출되지 않았습니다.")
```
